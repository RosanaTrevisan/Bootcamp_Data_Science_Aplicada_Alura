```
# Criando função para criar dicionário a partir do arquivo dicionario.xls
def cria_dicionario_pesquisa(lista_chaves, lista_valores):
  """
  Cria um dicionário com todos os códigos identificadores e as perguntas da pesquisa.

  Parâmetros:
  lista_chaves (list): lista contendo a coluna 'VARIÁVEL' do arquivo .xls
  lista_valores (list): lista contendo a coluna 'QUESTIONÁRIO DO ALUNO' do arquivo .xls

  Saída:
  dicionario (dict) contendo todas as perguntas da pesquisa no formato:  
  'REGEOGR': {1: 'Norte',
              2: 'Nordeste',
              3: 'Sudeste',
              4: 'Sul',
              5: 'Centro-Oeste',
              'pergunta': 'Região geográfica'
              }
  onde:
    'REGEOGR', usado como chave única no dicionário, contém o código da pergunta, 
                  conforme arquivo Dicionario_PENSE_Microdados_AMOSTRA2.xls;
    1 a 5 (podem variar conforme a pergunta) são as diferentes opções numéricas 
                  apresentadas para seleção da resposta à pergunta;
    'pergunta' é a chave que contém, como valor, a pergunta por extenso.
  """
  # Definindo variáveis
  dicionario = {}
  id_linha = 0
  variavel = lista_chaves
  dado = lista_valores
  ultima_chave = ''
  max_id = 1220

  # Lê a primeira chave
  chave = str(variavel[id_linha])
  # Alguns números vem com o tipo str, então converto tudo para str
  # para medir o comprimento do dado lido

  # Repetirá a execução enquanto chave não for vazia
  while((chave != "") and (id_linha < max_id)):
  # while(chave != ""):
    # print(f'chave: {chave} no início do while principal')
    # Verifica, pelo tamanho, se encontrou a chave de uma pergunta
    if (len(chave) > 2):
      # print(f'len(chave)>2 no primeiro if')
      # input(f'Continua?')
      # VB17001 e VB17002 tem formatos diferentes, então precisam ser tratadas
      # como exceções. VB17003 vem logo depois de VB17002, o que faz com que
      # ela precise ser tratada de forma diferente também
      if ((chave == 'VB17002') or (chave == 'VB17003')):
        # Cria a chave e a pergunta encontrada no dicionário
        dicionario[chave] = {'pergunta': dado[id_linha]}
        # print(f'{chave}: {dicionario[chave]}')
        ultima_chave = chave
        id_linha += 1
        chave = str(variavel[id_linha])
      elif ((ultima_chave == 'VB17001') or (ultima_chave == 'VB17002')):
        # print(f'ultima_chave: {ultima_chave}')
        # input(f'Continua?')
        dicionario[ultima_chave][chave] = dado[id_linha]
        # print(dicionario[ultima_chave])
        id_linha += 1
        chave = str(variavel[id_linha])
      else:
        # Cria a chave e a pergunta encontrada no dicionário
        dicionario[chave] = {'pergunta': dado[id_linha]}
        # print(f'{chave}: {dicionario[chave]}')
        ultima_chave = chave
        id_linha += 1
        try:
          chave = str(variavel[id_linha])
        except:
          break
        # Repetirá enquanto os dados lidos tiverem comprimento menor que 3
        while (len(chave) <= 2):
          # Os campos com as opções das respostas tem comprimento 1 ou 2
          # Exceto para as questões VB17001 e VB17002
          # print(f'chave_len<=2. chave: {chave}')
          # Criar chave secundária (aninhada) para os opções de resposta com
          # suas descrições
          dicionario[ultima_chave][int(chave)] = dado[id_linha]
          id_linha += 1
          chave = str(variavel[id_linha])
          # print(f'chave: {chave} no fim do while ds opções')
    else:
      id_linha += 1
      chave = str(variavel[id_linha])
      
  return dicionario


dic_pense = cria_dicionario_pesquisa(variavel, dado)
```

```
# Criando função para gerar o dicionário para uma pergunta específica
def cria_dicionario(dado_string, dicionario):
  """
  Cria um dicionário, para uma pergunta específica, com as possíveis opções de resposta e seus significados.

  Parâmetros:
  dado_string (str): código da pergunta para a qual será gerado o dicionário
  dicionario (dict): dicionário contendo todas as perguntas da pesquisa e seus códigos

  Saídas:
  dicionário no formato: {'dicionario': dic, 'descricao': descricao}
  onde:
    dic (dict) é um dicionário com as opções de resposta e seus significados
    descricao (str) é a pergunta (dado_string) por extenso.
  """
  # Acessando dados do dicionario
  dic = dicionario[dado_string].copy()
  descricao = dic['pergunta']
  dic.pop('pergunta')
  retorno = {'dicionario': dic, 'descricao': descricao}
  return retorno

# Definindo o dado a ser estudado
dado = 'REGEOGR'

# chamar função para criar o dicionário
resultado = cria_dicionario(dado, dic_pense)
resultado
```

```
#inserindo a coluna de IMC na base de dados
dados["IMC"] = dados["VB17003"]/ ((dados["VB17004"]/100)**2)
dados["IMC"]
```

```
#inserindo uma coluna para classificação dos dados IMC
classes = [0, 18.5, 25, 30, 60]
labels = ["abaixo", "normal", "sobrepeso", "obesidade"]

dados["CLASSE_IMC"] = pd.cut(x = dados["IMC"],
       bins = classes,
       labels = labels,
       include_lowest = True)

dados[["IMC", "CLASSE_IMC"]]
```

## IMC x Idade

```
ax = sns.boxplot(x = "VB01003", y = "IMC", data = dados)
ax.figure.set_size_inches(12, 6)
ax.set_title("Boxplot do IMC por Idade")
ax.set_ylabel("IMC")
ax.set_xlabel("Idade")
ax;
```

Aparentemente há uma tendência de aumento do IMC médio com a idade. Na sequência farei outro tipo de gráfico para confirmar, mas utilizando a Classe IMC.

## Idade x Classe IMC

```
idade_1 = dados['VB01003'].copy()
idade_1.rename('Idade', inplace=True)
IMC_1 = dados["CLASSE_IMC"].copy()
IMC_1.rename('CLASSE IMC', inplace=True)

idade_1_dic = cria_dicionario('VB01003', dic_pense)['dicionario']

frequencia_cruzada = pd.crosstab(idade_1, IMC_1, normalize = 'index') * 100
frequencia_cruzada = frequencia_cruzada.round(1)
frequencia_cruzada.rename(index = idade_1_dic, columns = IMC_1, inplace = True) 
frequencia_cruzada
```

```
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(frequencia_cruzada, linewidths=.5, ax=ax, cmap='coolwarm', annot=True)
plt.title('Idade x Classe IMC \n', fontsize=18)
plt.show()
```

Com o aumento da idade, observa-se diminuição da quantidade de jovens abaixo do peso e aumento de sobrepeso e obesidade.

## Idade x Classe IMC x Gênero

```
#separando as colunas que serão trabalhadas
sexo = dados['VB01001'].copy()
sexo.rename('Gênero', inplace=True)
idade_1 = dados['VB01003'].copy()
idade_1.rename('Idade', inplace=True)
IMC_1 = dados["CLASSE_IMC"].copy()
IMC_1.rename('CLASSE IMC', inplace=True)

#criando os dicionários
sexo_dic = cria_dicionario('VB01001', dic_pense)['dicionario']
idade_1_dic = cria_dicionario('VB01003', dic_pense)['dicionario']

# Trocando dados numéricos por dados categóricos
sexo_cat = sexo.map(sexo_dic)
idade_1_cat = idade_1.map(idade_1_dic)

#Tabela de frequência cruzada
frequencia_cruzada = pd.crosstab(index = [sexo_cat, idade_1_cat],
                                 columns= IMC_1, normalize = 'index')*100
frequencia_cruzada = frequencia_cruzada.round(1)
frequencia_cruzada
```

Quanto maior a classe IMC (mais gordos), maior o % de alunos que tomaram medicamentos para perder peso.
